```python
import numpy as np
import pickle
from DoS_estimation import dos_estimation
```

```python
with open('true_distribution_df_range51.pickle', 'rb') as p:
    true_distribution_df = pickle.load(p)
with open('rs_samples_diff_size_51.pickle', 'rb') as p:
    rsdf_list = pickle.load(p)
with open('es_samples_diff_size_51.pickle', 'rb') as p:
    esdf_list = pickle.load(p)
with open('es_f_diff_size_51.pickle', 'rb') as p:
    es_f_list = pickle.load(p)
```

```python
samples_num_idx = 4
best_hellinger_es_20000_idx = 0
best_hellinger_rs_20000_idx = 2

esdf = esdf_list[samples_num_idx]
esdf = esdf[esdf['trial']==best_hellinger_es_20000_idx]
rsdf = rsdf_list[samples_num_idx]
rsdf = rsdf[rsdf['trial']==best_hellinger_rs_20000_idx]
f_es = es_f_list[samples_num_idx][best_hellinger_es_20000_idx]

bnum = 10
betas = np.linspace(0.1,20,num=bnum)
```

```python
energy_max = 1
energy_min = 0
energy_level = 1000

de = dos_estimation(energy_max, energy_min, energy_level)
```

```python
true_distribution_df['log10_STY'] = np.log10(true_distribution_df['STY'])

energy_variable = 'Total_failure'
gray_variable = 'log10_STY'

energy_levels = [0,0.001,0.003,0.01,0.03,0.1,0.3]
gray_levels_num = 10
xlabel = "Space time yield (log10 kg/(m3*h)"
ylabel = "Failure rate (%)"
title = "True distribution of failure rate and space time yield (log10)"

de = dos_estimation(energy_max, energy_min, energy_level)
de.gray_plot(true_distribution_df, energy_variable, gray_variable, energy_levels, gray_levels_num, xlabel, ylabel, title)
```

```python
rsdf['log10_STY'] = np.log10(rsdf['STY'])
de.gray_plot(true_distribution_df, energy_variable, gray_variable, energy_levels, gray_levels_num, xlabel, ylabel, title, samples=rsdf)
```

```python
esdf['log10_STY'] = np.log10(esdf['STY'])
de.gray_plot(true_distribution_df, energy_variable, gray_variable, energy_levels, gray_levels_num, xlabel, ylabel, title, esdf, betas, f_es)
```
